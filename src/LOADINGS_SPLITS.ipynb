```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_source = '../results/loadings'
```

## COMPONENT 1

### GENES

```python
all_dfs_cp1_dna = []
all_dfs_cp2_dna = []
for i in range(5):
    df = pd.read_csv(os.path.join(data_source, 'LOADINGS_dna_' + str(i) + '.csv'), index_col=0)
    all_dfs_cp1_dna.append(df['comp1'])
    all_dfs_cp2_dna.append(df['comp2'])
```

```python
df_cp1_dna = pd.concat(all_dfs_cp1_dna, axis=1)
df_cp1_dna.columns = ['comp1_1', 'comp1_2', 'comp1_3', 'comp1_4', 'comp1_5']
df_cp1_dna = df_cp1_dna.fillna(0)
df_cp1_dna = abs(df_cp1_dna)
```

```python
df_cp1_dna['mean'] = df_cp1_dna.mean(axis=1)
df_cp1_dna = df_cp1_dna.sort_values(by=['mean'], ascending=False)
df_cp1_dna
```

```python
cols = ['tab:blue'] * 15
```

```python
plt.barh(df_cp1_dna.iloc[:15, -1].index, df_cp1_dna.iloc[:15, -1], color=cols)
```
